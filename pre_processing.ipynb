Now loading in the CSV file to explore and preprocess the data

```python
import pandas as pd

df = pd.read_csv('data/processed/tracker_data_filtered.csv')
df
```

```python
# Keeping only relevant columns
selected_columns = ['pid', 'cid', 'timestamp', 'device_id', 'trip_id', 
                    'latitude', 'longitude', 'Speed', 'distance', 'totalDistance', 'duration']
df_selected = df[selected_columns]
df_selected
```

```python
# Removing rows with NaN trip_id
df_selected = df_selected[~df_selected['trip_id'].isna()]
```

```python
# Checking how many trips with CID are associated with multiple device IDs
df = df_selected.copy()
cid_device_counts = df.groupby('cid')['device_id'].nunique().reset_index(name='device_count')

# Filter cids that have more than one unique device_id
cids_multiple_devices = cid_device_counts[cid_device_counts['device_count'] > 1]

# Count how many such cids exist
num_cids_multiple_devices = cids_multiple_devices.shape[0]

print(f"Number of cids with more than one associated device_id: {num_cids_multiple_devices}")

```

```python
# Creating a new idetntifier for each trip based on PID_DEVICEID_TRIPID
df_selected['timestamp'] = pd.to_datetime(df_selected['timestamp'])

df = df_selected.copy()

# Sort to ensure proper start/end detection
df = df.sort_values(by=['pid', 'device_id', 'trip_id', 'timestamp'])

# Compute start and end coordinates per trip
trip_summary = df.groupby(['pid', 'device_id', 'trip_id']).agg(
    start_lat=('latitude', 'first'),
    start_lon=('longitude', 'first'),
    end_lat=('latitude', 'last'),
    end_lon=('longitude', 'last')
).reset_index()

# Create a unique identifier string
trip_summary['trip_uid'] = trip_summary.apply(
    lambda x: f"{x.pid}_{x.device_id}_{x.trip_id}",
    axis=1
)


# Merge back to original dataset if needed
df = df.merge(trip_summary[['pid', 'device_id', 'trip_id', 'trip_uid']], 
              on=['pid', 'device_id', 'trip_id'], how='left')


df['trip_uid'].nunique()
```

```python
# Final cleaned dataframe with trip_id nulls removed and new trip_uid added
df.to_csv('data/processed/tracker_data_preprocessed.csv', index=False)
```
